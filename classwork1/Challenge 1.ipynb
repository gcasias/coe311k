```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
x=np.linspace(0,10,10)
plt.subplot(2,2,1)
plt.plot(x,np.sin(x),'r')

def f1(x,y):
    return (x*x+y*y)
x = np.linspace(-10, 10, 2000)
y = np.linspace(-10, 10, 2000)
X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
fig1 = plt.figure()
ax = plt.subplot(2,2,2,projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(90, 0)

plt.subplot(2,2,3)
y=np.sin(x)
plt.scatter(x,y)

plt.subplot(2,2,4)
plt.hist(2,3)

plt.show()
```
