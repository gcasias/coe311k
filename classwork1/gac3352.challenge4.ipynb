```python
import numpy as np

x = 0 

for n in range(5000): #First 5,000 terms
    x += (-1)**n / (2*n +1) #Leibniz Series by term
    
    if (n + 1) % 100 == 0: #Asked ChatGPT explaining prompt about every 100 terms, suggested to use remainder to do this
        approx_pi = 4 * x
        error = abs(approx_pi - np.pi) / np.pi
        print ("Terms:", n+1, "Relative Error:", error)
```
