```python
import numpy as np
import time
import matplotlib.pyplot as plt

# Was having a bit of trouble with my graph so Chat GPT suggested this
%matplotlib inline

def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

# Each matrix is 2x larger in rows and columns
sizes = [3,6,12,24,48,96,192,384]

naive_times = []
numpy_times = []

for n in sizes:
    print(f"Matrix size: {n}x{n}")
    
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)

    start = time.time()
    naive_matrix_multiplication(A,B)
    end = time.time()
    naive_times.append(end - start)
    
    start = time.time()
    np.dot(A, B)
    end = time.time()
    numpy_times.append(end - start)

plt.plot(naive_times, label="Naive")
plt.plot(numpy_times, label="Numpy")
plt.xlabel('Matrix Number')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```
