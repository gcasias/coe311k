```python
import numpy as np
import matplotlib.pyplot as plt

# Asked ChatGPT for help on how to begin code before approximations
f = lambda x:np.sin(x)

# Domain
a,b = 0,2 *np.pi
N = 50

x = np.linspace(a,b,N)

# Step size
h = (b - a) / (N - 1)

# Exact (cos(x))
exact = np.cos(x)

# Central Difference Approximation for the derivative
df_dx = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
forward = (f(x+h) - f(x)) / h

# Backward Difference Approximation
backward = (f(x) - f(x-h)) / h

# Plot the approximation and the true derivative
plt.figure(figsize=(10,6))
plt.plot(x, exact, label='cos(x)')
plt.plot(x, forward,'--', label='Forward Difference')
plt.plot(x, backward,'--',label='Backward Difference')
plt.plot(x,df_dx,'o--',label='Central Difference')
plt.xlabel('x')
plt.ylabel("Derivative")
plt.title("Comparison of Derivative Approximations")
plt.legend()
plt.show()
# Ran my code through ChatGPT like always to ensure no errors
```
