```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, y):
    return -1000*y + 3000 - 2000*np.exp(-t)

def forward_euler(f, y0, t0, tf, h):
    
    t_values = np.arange(t0, tf + h, h)
    y_values = np.zeros_like(t_values)
    y_values[0] = y0  
    
    for i in range(1, len(t_values)):
        y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])
        
    return t_values, y_values

def exact(t):
    return 3 - 0.998*np.exp(-1000*t) - 2.002*np.exp(-t)

y0 = 0
t0 = 0
tf = 30
h_values = [0.1, 0.01, 0.001]

plt.figure()

t_exact = np.linspace(t0, tf, 2000)
plt.plot(t_exact, exact(t_exact), label="Exact")

for h in h_values:
    t_values, y_values = forward_euler(f, y0, t0, tf, h)
    plt.plot(t_values, y_values, label=f"h={h}")

plt.xlabel('Time (t)')
plt.ylabel('y(t)')
plt.title('Forward Euler vs Exact Solution (Stiff ODE)')
plt.legend()
plt.grid(True)
plt.show()

# Ran my code through ChatGPT to ensure no errors and meets challenge requirements
```
