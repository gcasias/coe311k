```python
import numpy as np

# Data points
X = np.array([7,10,24,37,75])
Y = np.array([4.5,6.8,9.2,3.7,9.4])

# Stack the design matrix with a column of ones (for the inetrcept)
A = np.vstack([X,np.ones(len(X))]).T

print("X:")
print(X)

print("Y:")
print(Y)

print("A:")
print(A)
```

```python
# Solve using least squares
m,c = np.linalg.lstsq(A,Y,rcond=None)[0]
print(f"Slope:{m},Intercept:{c}")
```
