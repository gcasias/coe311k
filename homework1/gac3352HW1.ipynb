```python
import random
import numpy

def montecarlo_pi(number_of_shots): #Monte Carlo from class
    number_of_hits = 0
    for i in range(0,number_of_shots):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        
        if (x*x + y*y <1):
            number_of_hits = number_of_hits + 1
        
    return 4*number_of_hits/number_of_shots
```

```python
import math
import matplotlib.pyplot as plt

absolute_errors = [] # Asked Copilot why my plot wasn't working and it suggested to store & create a list
sample_sizes =[]

for i in range (10,1000,20): # Changed so it would be in steps of 20
    calculated_pi = montecarlo_pi(i)
    true_pi = numpy.pi
    true_error = abs(true_pi - calculated_pi)

    absolute_errors.append(true_error) # Store for list
    sample_sizes.append(i) # Store for list
    
print(i,calculated_pi,true_error)

plt.plot(sample_sizes,absolute_errors)
```
