```python
import numpy as np
import math
import matplotlib.pyplot as plt

def taylor_sin(x,n_terms):
    approximation = 0 
    num = x
    den = 1.0
    
    for n in range(n_terms):
        sin = (-1)**n
        term = sin * (num/den)
        approximation += term
        
        num = num*(x*x)
        den = den * (2*n +2) * (2*n +3)
       
    return approximation

x_values = np.arange(0,10.1,0.1)

plt.figure(figsize=(10,6))

for n in [1,2,3,4]:
    errors =[]
    for x in x_values:
        approx = taylor_sin(x,n)
        exact = np.sin(x)
        errors.append(abs(exact-approx))
        
    plt.plot(x_values,errors,label=f'(n) Term(s)'
             
#The trunction error increases as you move up in terms. The error does not decrease uniformly for all the terms.
```
