```python
import numpy as np
import matplotlib.pyplot as plt

# Create Random Linear Function
m_true = np.random.uniform(-5,5)
b_true = np.random.uniform(-5,5)

print("Original equation:")
print(f"y = {m_true}x + {b_true}")

# 0 to 1, steps of 0.01
X = np.arange(0,1.01,0.01)

# Calculated Y
noise = np.random.uniform(-1,1,len(X))
Y = m_true*X + b_true + noise

n = len(X)

# Followed the algorithim in slides
sum_x = np.sum(X)
sum_y = np.sum(Y)
sum_x2 = np.sum(X**2)
sum_xy = np.sum(X*Y)

m = (n*sum_xy - sum_x*sum_y)/(n*sum_x2 - sum_x**2)
c = (sum_y - m*sum_x)/n

print("\nCustom Least Squares:")
print(f"Slope: {m}")
print(f"Intercept: {c}")

# Numpy least Squares
A = np.vstack([X,np.ones(len(X))]).T
m_np,c_np = np.linalg.lstsq(A,Y,rcond=None)[0]

print("\nNumpy Least Squares:")
print(f"Slope: {m_np}")
print(f"Intercept: {c_np}")

Y_pred = m*X + c

# Residuals
residuals = Y - Y_pred

print("\nLargest residual:", np.max(residuals))
print("Smallest residual:", np.min(residuals))

plt.scatter(X,Y,label="Data Points")
plt.plot(X,Y_pred,label="Least Squares Line")
plt.legend()
plt.show()
```
