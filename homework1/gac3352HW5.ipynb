```python
import numpy as np
import matplotlib.pyplot as plt

X_rand = np.sort(np.random.rand(5)*5)
Y_rand = np.random.rand(5)*5

# My choice of m and b
m1 = 2
b1 = 3

X = np.arange(0,1.01,0.01)

# create Y with random noise
Y = m1*X + b1 + np.random.rand(len(X))

# sort X
idx = np.argsort(X)
X = X[idx]
Y = Y[idx]

# My Least Squares
n = len(X)

m = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y)) / (n*np.sum(X**2) - (np.sum(X))**2)
b = (np.sum(Y) - m*np.sum(X)) / n

print("Custom Least Squares")
print("m =",m)
print("b =",b)

# Numpy least Squares
A = np.vstack([X,np.ones(len(X))]).T
m_np, b_np = np.linalg.lstsq(A,Y,rcond=None)[0]

print("\nNumpy Least Squares")
print("m =",m_np)
print("b =",b_np)

# -------- Residuals --------
Y_pred = m*X + b
res = Y - Y_pred

print("\nLargest residual:",res[np.argmax(abs(res))])
print("Smallest residual:",res[np.argmin(abs(res))])


plt.scatter(X,Y)
plt.plot(X,Y_pred,label="Custom Fit")
plt.plot(X,m_np*X+b_np,label="Numpy Fit")
plt.legend()
plt.show()
```
