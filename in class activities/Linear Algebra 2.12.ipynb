Linear Algebra in Python

```python
import numpy as np
```

```python
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])
```

```python
print(v1)
print (v2)
```

```python
v_add = v1 + v2
v_sub = v1 - v2
```

```python
print(v_add)
print(v_sub)
```

```python
scalar = 2
v_scaled = scalar * v1
```

```python
print(v_scaled)
```

```python
dot_product = np.dot(v1,v2)
```

```python
print(dot_product)
```

```python
magnitude_v1 = np.linalg.norm(v1)
print(magnitude_v1)
```

```python
A = np.array([[1,2,3],[4,5,6]])
print(A)
print ()

B = np.array([[7,8,9],[10,11,12],[13,14,15]])
print(B)
print()
```

```python
C = np.array([[1,2,3],[4,5,6]])
matrix_add = A + C
matrix_sub = A - C

print("A+C=", matrix_add)
print("A-C=", matrix_sub)
```

```python
A = np.array ([[2,1],[1,2]])
e_value, e_vector = np.linalg.eig(A)
```

```python
print("eigenvalue:",e_value)
print("eigenvector:",e_vector)
```

```python
A = np.array([[4,7],[2,6]])
A_inv = np.linalg.inv(A)

print(A)
print(A_inv)
```

```python
identity = np.dot(A,A_inv)
print(identity)
```
